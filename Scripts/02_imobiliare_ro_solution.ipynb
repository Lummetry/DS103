```python
df = df.iloc[:,:-1]
```

```python
TARGET_COLUMN = 'Price'
```

```python
X_raw = df.values[:,:-2]
y = df[TARGET_COLUMN].values
X_raw.shape
```

```python
X = (X_raw - X_raw.mean(axis=0)) / X_raw.std(0)
X.mean()
```

```python
n_obs = X.shape[0]
prc_test = 0.15
```

```python
X[0]
```

```python
y[0]
```

```python
random_seed = 1010
```

```python
np.random.seed(random_seed)
```

```python
test_positions = np.random.choice(2, size=n_obs, p=[1 - prc_test, prc_test]) 
```

```python
X_train = X[~test_positions.astype(bool)]
y_train = y[~test_positions.astype(bool)]
X_test = X[test_positions.astype(bool)]
y_test = y[test_positions.astype(bool)]
print("X_train/y_train: {}/{}  X_test/y_test: {}/{}".format(X_train.shape, y_train.shape, X_test.shape, y_test.shape))
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_trn, X_tst, y_trn, y_tst = train_test_split(X, y, test_size=prc_test, random_state=random_seed)
print("X_trn/y_trn: {}/{}  X_tst/y_tst: {}/{}".format(X_trn.shape, y_trn.shape, X_tst.shape, y_tst.shape))
```

```python
Theta = np.linalg.pinv(X_train.T.dot(X_train)).dot(X_train.T).dot(y_train)
```

```python
Theta
```

```python
df.columns[:-2]
```

```python
df_explain = pd.DataFrame(
    {
        'Feature': df.columns[:-2],
        'Importance':Theta
    }
)
df_explain
```

```python
y_pred = X_test.dot(Theta)
```

```python
df_pred_vs_real = pd.DataFrame(
    {
        'Reality' : y_test,
        'Predictions' : y_pred
    }
)
df_pred_vs_real.head(20)
```
